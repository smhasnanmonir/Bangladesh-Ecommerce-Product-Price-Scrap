```python
import PyPDF2

def pdf_to_text_pypdf2(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text

# Usage
text = pdf_to_text_pypdf2('document.pdf')
print(text)
```
